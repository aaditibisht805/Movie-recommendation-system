```python
data1 = pd.read_csv('tmdb_5000_movies.csv')
data2=pd.read_csv('tmdb_5000_credits.csv')
```

```python
data1 = data1.merge(data2,on ='title')   # merging 2 dataframes on the basis of common column title
```

```python
data1.head(1)
```

```python
data1=data1[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
data1.head(1)
```

```python

#remove duplicate data and missing data
data1.isnull().sum()
```

```python
data1.dropna(inplace=True)
```

```python

#to convert string of list to list , avoid error('string indices must be integer')
import ast
```

```python
def convert(obj):
    list=[]
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
data1['genres'] = data1['genres'].apply(convert)
```

```python
data1.head(10)
```

```python
data1['keywords']=data1['keywords'].apply(convert)
```

```python
def convert1(obj):
    list=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            list.append(i['name'])
            count+=1
        else:
            break;
    return list
            
```

```python
data1['cast']=data1['cast'].apply(convert1)
```

```python
def find_director(obj):
    list=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            list.append(i['name'])
            break
    return list
```

```python
data1['crew']=data1['crew'].apply(find_director)
```

```python
#convert overview to list
data1['overview']= data1['overview'].apply(lambda x:x.split())
```

```python
data1.head(50)
```

```python
# to avoid name collisions between different column elements we need to remove spaces between them so that they becomes one word 
data1['genres'] = data1['genres'].apply(lambda x:[i.replace(" ","")for i in x])
data1['keywords'] = data1['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
data1['cast'] = data1['cast'].apply(lambda x:[i.replace(" ","")for i in x])
data1['crew'] = data1['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
data1.head(10)
```

```python
data1['tags']=data1['overview']+data1['genres']+data1['keywords']+data1['cast']+data1['crew']
```

```python
data1.head(50)
```

```python
new_data1=data1[['movie_id','title','tags']]
```

```python
new_data1
```

```python
new_data1['tags']=new_data1['tags'].apply(lambda x:" ".join(x))   
#converting tags to string again from the list
```

```python
new_data1['tags'][0]     # all the dataframe combined and are converted to string para
```

```python
new_data1['tags']=new_data1['tags'].apply(lambda x:x.lower()) 
```

```python
# now to convert the text to vectors so that there are neary 5000(dataset size)vectors in space 
#technique used bag of words 
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# remove stop words 
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_data1['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()  # most frequent words 
```

```python
#stemming-- similar words replaced to same word 
!pip install nltk
```

```python
new_data1
```

```python
from nltk.stem.porter import PorterStemmer
p=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(p.stem(i))
    return " ".join(y)
    
```

```python
new_data1['tags']=new_data1['tags'].apply(stem)
```

```python
#higher dimensions cosine distance 
from sklearn.metrics.pairwise import cosine_similarity   
#1-more similiar
#0- less similar
```

```python
similarity= cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:11]
```

```python
def recommend(movie):
    movie_index=new_data1[new_data1['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list= sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:11]
    for i in movie_list:
        print(new_data1.iloc[i[0]].title)
       
         
    
```

```python
recommend('Avatar')
```
